```python
import pandas as pd
```

## 1. Coleta do nome e a sede de todos os times que disputaram o Campeonato Paulista em 2022

### Paulistão

```python
tables = pd.read_html('https://pt.wikipedia.org/wiki/Campeonato_Paulista_de_Futebol_de_2022')
```

```python
a1 = tables[4].drop(['Em 2021', 'Estádio', 'Material Esportivo', 'Títulos'], axis=1)
a1.head()
```

```python
a1.shape
```

### A2

```python
tables = pd.read_html('https://pt.wikipedia.org/wiki/Campeonato_Paulista_de_Futebol_de_2022_-_S%C3%A9rie_A2')
```

```python
a2 = tables[4].drop(['Em 2021', 'Estádio', 'Capacidade', 'Títulos'], axis=1)
a2.head()
```

```python
# Ajuste no nome do time
a2.iloc[2]['Equipe'] = 'Lemense Futebol Clube'
```

```python
a2.shape
```

### A3

```python
tables = pd.read_html('https://pt.wikipedia.org/wiki/Campeonato_Paulista_de_Futebol_de_2022_-_S%C3%A9rie_A3')
```

```python
a3 = tables[3].drop(['Em 2021', 'Estádio', 'Capacidade', 'Participações', 'Títulos'], axis=1)
a3.head()
```

```python
a3.shape
```

### Segunda Divisão

```python
tables = pd.read_html('https://pt.wikipedia.org/wiki/Campeonato_Paulista_de_Futebol_de_2022_-_Segunda_Divis%C3%A3o')
```

```python
sd = tables[3].drop(['Última participação'], axis=1)
sd.head()
```

```python
# Ajuste no nome da coluna
sd.rename(columns={'Clube': 'Equipe'}, inplace=True)
```

```python
# Adicionar a divisão de cada time
a1['Série Atual'] = 'A1'
a2['Série Atual'] = 'A2'
a3['Série Atual'] = 'A3'
sd['Série Atual'] = 'Segunda Divisão'
```

```python
# Agrupando todos os times que jogaram uma das quatro divisões estaduais em 2022
paulista = pd.concat([a1, a2, a3, sd])
paulista.shape
```

## 2. Coleta do PIB, IDH e número de habitantes dos munícipios paulistas

### PIB

```python
tables = pd.read_html('https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_S%C3%A3o_Paulo_por_PIB')
```

```python
pib_sp = tables[1]
pib_sp.rename(columns={'Unnamed: 0': 'Pos', 'PIB (2018)em milhões': 'PIB', 'Município': 'Cidade'}, inplace=True)
pib_sp.head()
```

### IDH

```python
tables = pd.read_html('https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_S%C3%A3o_Paulo_por_IDH-M')
```

```python
idh_sp = tables[1]
idh_sp.columns = ['Pos', 'Cidade', 'IDH', 'IDH-renda', 'IDH-longevidade', 'IDH-educação']
idh_sp.head()
```

### Número de habitantes

```python
tables = pd.read_html('https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_S%C3%A3o_Paulo_por_popula%C3%A7%C3%A3o_(2010)')
```

```python
pop_sp = tables[1].rename(columns={'Unnamed: 0': 'Pos', 'Nome': 'Cidade', 'Pop. Mun.': 'População'})

# Removendo distritos, vilas, etc
pop_sp = pop_sp.dropna(subset=['População'])

pop_sp.drop(['Pop. Distr.', 'Pop. Urb.', 'Categoria'], axis=1, inplace=True)
pop_sp.head()
```

## 3. Concatenando times e dados dos munícipios 

### Times x PIB

```python
time_pib = paulista.merge(pib_sp, on='Cidade')
time_pib.head()
```

```python
# Função auxiliar para converter o campo PIB de string para decimal
def monetaryValue(x):
    x = x.replace('R$ ', '')
    x = x.replace('.', '')
    x = x.replace(',', '.')
    return float(x)
```

```python
time_pib['PIB'] = time_pib['PIB'].apply(monetaryValue)
time_pib.head()
```

### Times x IDH

```python
time_idh = paulista.merge(idh_sp, on='Cidade')
time_idh.head()
```

### Times x População 

```python
time_pop = paulista.merge(pop_sp, on='Cidade')
time_pop.head()
```
